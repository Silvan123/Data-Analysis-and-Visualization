```python
import pandas as pd
import pprint as pp
import plotly.plotly as py
import plotly.graph_objs as go
import datetime as datetime

df = pd.read_csv('../data/AllGunShootings.csv') 



def get_axes():    
    # insert casualties(casualties AND injured) per incident
    df.insert(value = [kill + inj for kill, inj in zip(df['n_killed'], 
              df['n_injured'])], loc= df.shape[1], column = 'casualties')
    
    # set start values and constant
    old_day = 1
    casualties_num = 0
    incidents_num = 0
    incidents = []
    casualties = []
    rows = df.shape[0]
    
    weight29 = 1/50
    weight30 = 1/46
    weight31 = 1/30
    standardweight = 1/51
    
    
    # iterate over rows
    for i in range(278, rows):
        
        # increase if month is the same
        if int(df.iloc[i, 2][8:]) == old_day:
            casualties_num += df.iloc[i, 19]
            incidents_num += 1
        # gather data if new month is there
        elif(len(incidents) != 31):
            if old_day == 29:
                casualties.append(weight29 * casualties_num)
                incidents.append(weight29 * incidents_num)
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
            elif old_day == 30:
                casualties.append(weight30 * casualties_num)
                incidents.append(weight30* incidents_num)
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
            elif old_day == 31:
                casualties.append(weight31 * casualties_num)
                incidents.append(weight31* incidents_num)
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
            else:
                casualties.append(standardweight * casualties_num)
                incidents.append(standardweight * incidents_num)
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
                
        else:
            if old_day == 29:
                casualties[old_day - 1] += weight29 * casualties_num
                incidents[old_day - 1] += weight29 * incidents_num
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
            elif old_day == 30:
                casualties[old_day - 1] += weight30 * casualties_num
                incidents[old_day - 1] += weight30 * incidents_num
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
            elif old_day == 31:
                casualties[old_day - 1] += weight31 * casualties_num
                incidents[old_day - 1] += weight31 * incidents_num
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
            else:
                casualties[old_day - 1] += standardweight * casualties_num
                incidents[old_day - 1] += standardweight * incidents_num
                old_day = int(df.iloc[i, 2][8:])
                inj_num = 0
                casualties_num = 0
                incidents_num = 0
    
    # append vals for last month
    casualties[old_day - 1] += weight31 * casualties_num
    incidents[old_day - 1] += weight31 * incidents_num
    return [casualties, incidents]

months = [date for date in range(1, 32)]

axes = get_axes()

trace0 = go.Scatter(
    x= months,
    y = axes[1],
    name = 'incidents',
    line = dict(
        shape = 'spline'
    )
)

trace1 = go.Scatter(
    x = months ,
    y = axes[0],
    name = 'casualties',
    line = dict(
        color = 'rgb(255, 127, 14)',
        shape = 'spline'
    )
)

layout0 = dict(title = 'Average amount of incidents per day of the month',
              xaxis = dict(title = 'Month'),
              yaxis = dict(title = 'Incidents'),
              )

layout1 = dict(title = 'Average amount of casualties per day of the month',
              xaxis = dict(title = 'Month'),
              yaxis = dict(title = 'Casualties'),
              )
fig = dict(data = data0, layout = layout0)
data0 = [trace0]
data1 = [trace1]

py.iplot(fig, filename='data_month')
```
