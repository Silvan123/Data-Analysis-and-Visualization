```python
import pandas as pd
import pprint as pp


df = pd.read_csv('../data/AllGunShootings.csv') 
dfList = df['participant_age_group'].fillna('')
dfList2 = df['participant_type'].fillna('')
dfList3 = df['incident_characteristics'].fillna('')
### Set string from csv to dict datastructure, BEWARE: all keys are of 
### type string






# get all dicts in Lists
age=[to_dict(x) for x in dfList]
part=[to_dict(x) for x in dfList2]

#counters
teen_suspect=0
mass=0
teens_mass=0

# iterate over the lists and compare the values of each dict to match the teens to subject/suspects
# then check whether the corresponding shooting was a mass shooting
for index in range(len(age)):
    for key, value in age[index].items():
        if 'Subject-Suspect' in part[index][key] : 
            if 'Teen 12-17' in age[index][key] :      
                teen_suspect += 1
                if 'Mass Shooting' in dfList3[index]:
                    teens_mass += 1
                
#number of mass shootings
for index in range(len(dfList3)):
    if 'Mass Shooting' in dfList3[index]:
        mass += 1            


#comparing numbers
total = len(dfList3)
non_teens = total-teen_suspect
non_teens_mass = mass-teens_mass
percent_non_teen = (non_teens_mass/non_teens)
percent_teen = (teens_mass/teen_suspect)
comparison = (percent_teen/percent_non_teen)


d = {'non-teen suspects': [non_teens], 'teen suspects': [teen_suspect]}
d2= {'non-teen suspects': [non_teens_mass], 'teen suspects': [teens_mass]}
df2 = pd.DataFrame(data=d)
df3 = pd.DataFrame(data=d2)
ax=df2.plot.barh(stacked=True);
ax2 = df3.plot.barh(stacked=True);
ax.set_xlabel("shootings")
ax2.set_xlabel("mass-shootings")


print('shootings with teens as Subject-Suspects: %d'% teen_suspect)
print('total mass shootings: %d'% mass)
print('mass shootings by teens: %d'% teens_mass)
print('teens are %f times more likely to commit mass shootings rather than other shootings in comparison to adults'% comparison)
```
