```python
import pandas as pd
import pprint as pp

df = pd.read_csv('../data/AllGunShootings.csv') 
dfList = df['participant_status'].fillna('')

### Set string from csv to dict datastructure, BEWARE: all keys are of 
### type string
def to_dict(string_dict):
    
    # set start values
    goal = {}
    key = ''
    val = ''
    len_string_dict = len(string_dict)
    key_switcher = True
    
    # Iterate over string indices
    for i in range(len_string_dict):
        
        # look for token
        if string_dict[i:i+2] == '::':
            key_switcher = False
        # look for seperator
        elif string_dict[i:i+2] == '||':
            key_switcher = True
            goal[key] = val
            val = ''
            key = ''
        # look for key
        elif key_switcher == True and (not string_dict[i] == ':' and not string_dict[i] == '|'):
            key += string_dict[i]
        # look for value
        elif not string_dict[i] == ':' and not string_dict[i] == '|':
            val += string_dict[i]
            
    # only add to dict if string is not empty
    if len_string_dict > 0:
        goal[key] = val
        
    return goal

# get all dicts in List
[to_dict(x) for x in dfList]
```
